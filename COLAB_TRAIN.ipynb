```python
# Import necessary libraries
import pandas as pd
import numpy as np
import torch
from torch import nn
from transformers import BertTokenizer, BertModel
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from tqdm import tqdm
import pickle
import os
import matplotlib.pyplot as plt


# Check for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

file1 = "data/books_all(1-5).csv"
file2 = "data/books_all(6-10).csv"
file3 = "data/books_all(11-15).csv"
file4 = "data/books_all(16-20).csv"
file5 = "data/books_all(21-25).csv"
file6 = "data/books_all(26-30).csv"
# Đọc từng file vào DataFrame
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
df5 = pd.read_csv(file5)
df6 = pd.read_csv(file6)
# Gộp hai DataFrame lại
df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
print(f"Dataset loaded with {df.shape[0]} games and {df.shape[1]} features")

# --- Data Preprocessing ---
# Fill missing values in 'Detailed description'
df['description'] = df['description'].fillna("")

# Function to flatten list-like data
def flatten_list(lst):
    if isinstance(lst, str):
        return lst.split(',')
    elif isinstance(lst, list) and all(isinstance(x, str) for x in lst):
        return lst
    elif isinstance(lst, list) and len(lst) == 1 and isinstance(lst[0], list):
        return lst[0]
    return [str(lst)]

# Standardize Genres, Tags, Categories
for col in ['genres', 'shelves']:
    df[col] = df[col].fillna('Unknown').apply(flatten_list)

# --- Feature Extraction ---
# Encode Genres, Tags, Categories
mlb_genres = MultiLabelBinarizer()
genres_matrix = mlb_genres.fit_transform(df['genres']).astype(np.float64)

mlb_shelves = MultiLabelBinarizer()
shelves_matrix = mlb_shelves.fit_transform(df['shelves']).astype(np.float64)
# Boost important tags

# BERT embeddings for Detailed description
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased').to(device)

def get_bert_embedding_batch(texts, batch_size=16):
    embeddings = []
    for i in tqdm(range(0, len(texts), batch_size), desc="Generating BERT embeddings"):
        batch = [str(text) if text is not None else "" for text in texts[i:i+batch_size]]
        inputs = tokenizer(batch, return_tensors='pt', truncation=True, padding=True, max_length=128).to(device)
        with torch.no_grad():
            outputs = bert_model(**inputs)
        embeddings.append(outputs.last_hidden_state.mean(dim=1).cpu().numpy())
    return np.vstack(embeddings)

desc_embeddings = get_bert_embedding_batch(df['description'].tolist())

# Combine features
features = np.hstack([genres_matrix, shelves_matrix, desc_embeddings])
print(f"Features shape: {features.shape}")
features = torch.tensor(features, dtype=torch.float32).to(device)

# --- Define and Train Autoencoder ---
class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 512), nn.BatchNorm1d(512), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(512, 256), nn.BatchNorm1d(256), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(256, 128), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(128, 256), nn.BatchNorm1d(256), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(256, 512), nn.BatchNorm1d(512), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(512, input_dim), nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Initialize model
input_dim = features.shape[1]
model = Autoencoder(input_dim=input_dim).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Training with mixed precision
from torch.cuda.amp import GradScaler, autocast
scaler = GradScaler()
losses = []  # Store losses for plotting

for epoch in range(50):
    model.train()
    epoch_loss = 0
    for i in range(0, len(features), 128):
        batch = features[i:i+128]
        optimizer.zero_grad()
        with autocast():
            output = model(batch)
            loss = criterion(output, batch)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        epoch_loss += loss.item()
    avg_loss = epoch_loss / (len(features) // 128 + 1)
    losses.append(avg_loss)
    print(f"Epoch {epoch+1}/50, Loss: {avg_loss:.4f}")

# Plot training loss
plt.plot(range(1, 51), losses, label='Training Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Autoencoder Training Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join('training_loss.png'))
plt.show()

# Extract latent representations
model.eval()
with torch.no_grad():
    latent_reps = model.encoder(features).cpu().numpy()

# --- PCA Comparison ---
pca = PCA(n_components=128)
pca_features = pca.fit_transform(features.cpu().numpy())
reconstructed = pca.inverse_transform(pca_features)
mse_pca = ((features.cpu().numpy() - reconstructed) ** 2).mean()
print(f"PCA MSE: {mse_pca:.4f}")

# Compare cosine similarity for first 5 games
autoencoder_sim = cosine_similarity(latent_reps[:5])
pca_sim = cosine_similarity(pca_features[:5])
print(f"Autoencoder Avg Similarity: {autoencoder_sim.mean():.2f}")
print(f"PCA Avg Similarity: {pca_sim.mean():.2f}")

# --- Generate Recommendations ---
# Example: Top 5 recommendations for the first game
similarities = cosine_similarity(latent_reps[0:1], latent_reps)[0]
top_indices = similarities.argsort()[-6:-1][::-1]  # Top 5 excluding self
top_similarities = similarities[top_indices]

print("Top 5 similar books for", df['title'].iloc[0])
for idx, sim in zip(top_indices, top_similarities):
    print(f"Books: {df['title'].iloc[idx]}, Similarity: {sim:.2f}")

# Save results
torch.save(model.state_dict(), os.path.join('models/autoencoder_model.pth'))
np.save(os.path.join('latent_reps.npy'), latent_reps)
with open(os.path.join('mlb_genres.pkl'), 'wb') as f:
    pickle.dump(mlb_genres, f)
with open(os.path.join('mlb_shelves.pkl'), 'wb') as f:
    pickle.dump(mlb_shelves, f)
df.to_csv(os.path.join('books_data_processed.csv'), index=False)

print("Model and components saved to Google Drive!")
```
